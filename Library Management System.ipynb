# Library Management System using Oops

```python
class Library:
    def __init__(self,name,book):
        self.name=name
        self.books=book
        self.lendBooks={}
        print(f"{self.name} library successfully created..")
        
    def display(self):
        print(f"We have these books available in {self.name}")
        print("Books \t\tCount")
        print("----------------------")
        for key_value in self.books.items():
            print(key_value[0],"  \t", key_value[1])
    
    def lendBook(self,book,user):
        
        if (book in self.books) and (self.books[book])==0:
            print("The book is taken by someone.")
            
        elif (book in self.books) and (book not in self.lendBooks):
            self.lendBooks.update({book:[user]})
            self.books[book]-=1
            print(f"{user}, Now you can take {book} ")
            
        elif (book in self.lendBooks) :
            if user in self.lendBooks.get(book):
                print(f"{user}, You have already taken {book}")
            else:
                self.lendBooks[book].append(user)
                self.books[book]-=1
                print(f"{user}, Now you can take {book} ")
            
        else:
            print("The book is not present in library")
        
    
    def addBook(self,book,count):
        if book in self.books:
            self.books[book]+=count
        else:
            self.books.update({book:count})
        print(f"The {book} has been added to the library")
    
    def issuedBooks(self):
        if self.lendBooks:    
            print(f"The Issued by {self.name}")
            for key_value in self.lendBooks.items():
                print(f"{key_value[0]} is taken by {', '.join(key_value[1])}")
        else:
            print(f"Not any book has issued by {self.name}")
        
    def returnBook(self,book,user):
        if (book in self.lendBooks) and (user in self.lendBooks[book]): 
            if len(self.lendBooks[book])>1:
                self.lendBooks[book].remove(user)
                self.books[book]+=1 
            else:    
                self.lendBooks.pop(book)
                self.books[book]+=1
            print("Returned succesfully")
        else:
            if book not in self.lendBooks:
                print(f"The {book} is not issued by {self.name}")
            else:
                print("User name is Invalid..")
            
    
```

```python
library=Library("Mumbai Coding Club",{'Physics':3,'Chemistry':2,'Biology':2,'Mathmatics':3,'History':4,"English":2})



while True:
    print("-----------------------------------------")
    print(f"Welcome to the {library.name} Library")
    print("1: Display Books")
    print("2: Lend a Book")
    print("3: Add a Book")
    print("4: Return a book")
    print("5: Issued Books")
    print("6: Exit the library\n")
    option=int(input("Select the above option:- \n"))
    if option not in [1, 2, 3, 4, 5]:
        print("Enter Valid option ")
    else:
        option=int(option)
        
    if option==1:
        library.display()
        
    elif option==2:
        book=input("Enter book name: ")
        user=input("Enter your name: ")
        library.lendBook(book,user)
        
    elif option==3:    
        book=input("Enter the book name you want to add: ")
        count=int(input("Eneter the number of book: "))
        library.addBook(book,count)
        
    elif option==4:
        book=input("Enter the book you want to return:")
        user=input("Enter your Name: ")
        library.returnBook(book,user)
        
    elif option==5:
        library.issuedBooks()
        
    elif option==6:
        print("Thanks for visiting the")
        break
```
